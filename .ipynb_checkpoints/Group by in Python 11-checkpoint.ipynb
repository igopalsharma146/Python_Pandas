# Group by in Python

```python
import pandas as pd
d1=pd.DataFrame({"name":["a","b","c","d","a","b","a","b","a","c","c","d"],
                 "S_1":[12,11,10,13,15,17,23,24,25,27,26,32],
                 "S_2":[12,13,14,45,35,36,63,45,33,63,34,42]})
d1
```

```python
#Grouping
var_new=d1.groupby("name")
print(var_new)
```

```python
#showing the our data
for x,y in var_new:
    print(x)
    print(y)
    print("\n")
```

```python
# getting perticular group data
var_new.get_group("a")
```

```python
# getting perticular group data
var_new.get_group("b")
```

```python
# getting minimum , maximum data
var_new.min()
```

```python
# getting minimum , maximum data
var_new.max()
```

```python
# getting minimum , maximum data
var_new.count()
```

```python
# getting minimum , maximum data
var_new.sum()
```

```python
# getting minimum , maximum data
var_new.mean()
```
