```python
#Arithmetic operations

#adding two matrix
print("DataFrame")
import pandas as pd
var=pd.DataFrame({"A":[3,4,5,6],"B":[7,6,5,9]})
print(var)
print("\n")

print("Adding Other Column")
var["c"]=[6,7,9,4]
print(var)
print("\n")

print("#adding two matrix")
var["d"]=var["A"]+var["B"]
print(var)
print("\n")

print("#asubtracting two matrix")
var["e"]=var["A"]-var["B"]
print(var)
print("\n")

print("#multipling two matrix")
var["f"]=var["A"]*var["B"]
print(var)
print("\n")

print("#division two matrix")
var["g"]=var["A"]/var["B"]
print(var)
print("\n")

print("#Printing Perticular column")
print(var["c"])
print("\n")

# performing other arithmetic operation
var["python"]=var["A"]<5
var["python_1"]=var["B"]>7
var["python_2"]=var["c"]<5
print(var)
```
