# Pivot Table and Melt Function in Pyhton pandas

```python
import pandas as pd
var=pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var
```

```python
#Reshape
pd.melt(var)
```

```python
# hum ya per kisi bhi column ko as a id bhi le sakte hai
pd.melt(var,id_vars=["days"])
```

```python
# changing the variale name
pd.melt(var,id_vars=["days"],var_name="python")
```

# pivot()

```python
import pandas as pd
var=pd.DataFrame({"days":[1,2,3,4,5,6],"St_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var
```

```python
var.pivot(index="days",columns="St_name")
```

```python
#getting the data of perticular column
var.pivot(index="days",columns="St_name",values="eng")
```

```python
import pandas as pd
var1=pd.DataFrame({"days":[1,1,1,1,2,2],"St_name":["a","b","a","a","b","b"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var1
```

```python
var1.pivot_table(index="St_name",columns="days")
```

```python
var1.pivot_table(index="St_name",columns="days",aggfunc="mean")
```

```python
var1.pivot_table(index="St_name",columns="days",aggfunc="sum")
```

```python
#margin
var1.pivot_table(index="St_name",columns="days",aggfunc="mean",margins=True)
```
