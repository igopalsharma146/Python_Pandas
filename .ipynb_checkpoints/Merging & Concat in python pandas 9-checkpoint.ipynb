# Merging & Concat in python pandas

```python
import pandas as pd

var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,4],"C":[0,1,2,3]})

#NOTE:agar hame two dataFrame ko merge karana hai to unke column ka name same hona chahiye.
pd.merge(var1,var2)
```

```python
import pandas as pd

var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[0,1,2,3]})

pd.merge(var1,var2,how="left")
```

```python
import pandas as pd

var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[0,1,2,3]})

pd.merge(var1,var2,how="right")
```

```python
import pandas as pd

var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[0,1,2,3]})

pd.merge(var1,var2,how="outer")
```

```python
# if we want to see absent and present data then we can use here indicator parameter
import pandas as pd

var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[0,1,2,3]})

pd.merge(var1,var2,how="outer",indicator=True)
```

```python
import pandas as pd

var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,5],"B":[0,1,2,3]})

pd.merge(var1,var2)
#here our data is not showing because we are giv the same column name in both variables or different values. 
#if in this condition we want to show the our data then we can use here left_index,right_index
pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
import pandas as pd

var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,5],"B":[0,1,2,3]})

pd.merge(var1,var2)

# suffixes ki help se hum hamare data ki value ko bhi change kar sakte hai,
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","Python"))
```

# conact

```python
sr1=pd.Series([1,2,3,4])
sr2=pd.Series([11,12,13,14])

pd.concat([sr1,sr2])
```

```python
import pandas as pd

d1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[0,1,2,3]})

pd.concat([d1,d2])
```

```python
import pandas as pd

d1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[0,1,2,3]})

pd.concat([d1,d2],axis=1)
```

```python
import pandas as pd

d1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
d2=pd.DataFrame({"A":[1,2],"B":[0,1]})

pd.concat([d1,d2],axis=1)
```

```python
import pandas as pd

d1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
d2=pd.DataFrame({"A":[1,2],"B":[0,1]})

#join parameter
pd.concat([d1,d2],axis=1,join="inner")
```

```python
import pandas as pd

d1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
d2=pd.DataFrame({"A":[1,2],"B":[0,1]})

#join parameter
pd.concat([d1,d2],axis=1,join="outer")
```

```python
import pandas as pd

d1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
d2=pd.DataFrame({"A":[1,2,3,5],"C":[0,1,2,3]})

#key ke according data merging
pd.concat([d1,d2],axis=1,keys=["d1","d2"])
```

```python
import pandas as pd

d1=pd.DataFrame({"A":[1,2,3,4]})
d2=pd.DataFrame({"A":[1,2,3,5],"C":[0,1,2,3]})

#key ke according data merging
pd.concat([d1,d2])
```

```python
import pandas as pd

d1=pd.DataFrame({"A":[1,2,3,4]})
d2=pd.DataFrame({"A":[1,2,3,5],"C":[0,1,2,3]})

pd.concat([d1,d2],axis=1)
```

```python
import pandas as pd

d1=pd.DataFrame({"A":[1,2,3,4]})
d2=pd.DataFrame({"B":[1,2,3,5],"C":[0,1,2,3]})

#key ke according data merging
pd.concat([d1,d2],axis=0)
```
