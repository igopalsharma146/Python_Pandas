# dropna & fillna



```python
#handling missing values
import pandas as pd
var=pd.read_csv("null.csv")
print(var)
```

```python
#yah raw ke alog work karta hai
var.dropna()
```

```python
#if we are write here axix=1 , then yah colom ke along work karega
var.dropna(axis=1)
```

```python
#agar hum chahte hai ki jis raw me all value null hai , sirf wahi row delete ho , In these situation we can use the 'how' parameter
var1=pd.read_csv("null1.csv")
print(var1)
```

```python
var1.dropna(how="all")
```

```python
#any parameter lene se jitne bhi row me null value hai, vo sabhi raw delete ho jayengi.
var1.dropna(how="any")
```

```python
#hum subset parameter ki help se perticular column ki null value ko remove kar sakte hai.
var1.dropna(subset=["Student"])
```

```python
#inplace parameter
# d=var1.dropna(inplace=True)
# print(d)

var1.dropna(inplace=True)
```

```python
#thress parameter ki help se bhi hum null values ko handle kar sakte hai.
var.dropna(thresh=1)
```

# Fillna

```python
#fillna
var.fillna("python")
```

```python
#agar hame perticular column ke along null value dalni hai,to hum uske liye dictionary ka use kar sakte hai.
var.fillna({"Income":"high","Student":"No","Credit_rating":"Fair"})
```

```python
#hum yaha per null value ko previous data or Backward data se bhi fill kar sakte hai.
print(var)
var.fillna(method="ffill") #forward filling
```

```python
 #backward filling
var.fillna(method="bfill")
```

```python
#fill along the axis
var.fillna(method="bfill",axis=0)
# var.fillna(method="ffill",axis=0)
```

```python
var.fillna(method="bfill",axis=1)
# var.fillna(method="ffill",axis=1)
```

```python
var.fillna("python",limit=2)
```

```python
#using inplace parameter
var.fillna(12,inplace=True)
print(var)
```
