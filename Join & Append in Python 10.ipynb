# Join & Append in Python

```python
import pandas as pd
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"C":[21,22,23,24],"D":[31,32,33,34]})

d1.join(d2)
```

```python
import pandas as pd
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"C":[21,22],"D":[31,32]})

#join method unmatch data ke sath bhi work karta hai.
d1.join(d2)
```

```python
import pandas as pd
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
d2=pd.DataFrame({"C":[21,22,23,24],"D":[31,32,33,34]},index=["a","b","c","d"])

d1.join(d2)
```

```python
import pandas as pd
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
d2=pd.DataFrame({"C":[21,22,23,24],"D":[31,32,33,34]},index=["a","b","c","d"])

d2.join(d1)
```

```python
import pandas as pd
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
d2=pd.DataFrame({"C":[21,22],"D":[31,32]},index=["a","b"])

print(d2.join(d1))

# using how parameter
d2.join(d1,how="right")
```

```python
import pandas as pd
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
d2=pd.DataFrame({"C":[21,22],"D":[31,32]},index=["a","b"])

# using how parameter
d2.join(d1,how="outer")      #here outer works as union and inner works as intersection
```

```python
import pandas as pd
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
d2=pd.DataFrame({"C":[21,22],"D":[31,32]},index=["a","b"])

# using how parameter
d2.join(d1,how="inner")      #here outer works as union and inner works as intersection
```

```python
import pandas as pd
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
d2=pd.DataFrame({"C":[21,22],"B":[31,32]},index=["a","b"])

# using how parameter
# d2.join(d1,how="inner")      #this will be throw an error because here column name is same

# by using the suffix we can change here column name at run time
d2.join(d1,how="outer",lsuffix="_12")
```

```python
import pandas as pd
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
d2=pd.DataFrame({"C":[21,22],"B":[31,32]},index=["a","b"])

# using how parameter
# d2.join(d1,how="inner")      #this will be throw an error because here column name is same

# by using the suffix we can change here column name at run time
d2.join(d1,how="outer",rsuffix="_12")
```
