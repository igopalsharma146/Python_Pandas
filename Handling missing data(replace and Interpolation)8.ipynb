# Handling missing data(replace and Interpolation)

```python
#replace
import pandas as pd
var=pd.read_csv("data.csv")
print(var.head())
```

```python
# replace
var.replace(to_replace="medium",value="Medium")
```

```python
var.replace(["Yes","No"],"No Idea")
```

```python
#hum regex parameter ke throw bhi replace kar sakte hai.
var.replace("[A-z]","Gopal",regex=True)
```

```python
# dictionary ke through replace
var.replace({"Age":"[A-z]"},22,regex=True)
```

```python
print(var)
var.replace("low",method="bfill") #isi tarah hum forward fill bhi laga sakte hai.
```

```python
# hum yaha per limit parameter ka use karke, limited data ko bhi replace kara sakte hai.
d=var.replace("low",method="ffill",limit=1)
print(d)
# print(var)
# agar hum chahte hai ki yaha per hamare data ki copy na bane
# or hamare original data ke ander bhi replacement ho jaye to uske liye hum inplace parameter ka use kar sakte hai.
# inplace= True karne se hamare original data me change ho jayega
```

# Interpolate

```python
import pandas as pd
var1=pd.read_csv("null2.csv")
print(var1)
```

```python
# interpolate non value ko fill karne ke hi kaam aata hai.
var1.interpolate()  #interpolate ke liye ,atleast one column numeric dtype ka hona chahiye.
#agar hum ise direct use karte hai to yah number ke data ke sath hi kaam karega.
```

```python
# intepolate linealy data ko fill karta hai
#agar hum chate hai ki linearly data filled na ho to uske liye hum yaha per methods ka use kar sakte hai.
#method:{linear,time,index,values,nearest,zero,slinear,quadratic,cubic,barycentric,polynomial,spline,piecewise_polynomial,from_derivatives,pchip,akima}
var1.interpolate(method="linear",axis=0)
#hum yaha per axis parameter ka bhi use kar sakte hai
#axis=0 raw ke along work karti hai.
#axis=1 column ke along work karti hai.
```

```python
# var1.interpolate(method="linear",axis=1)           #this will be throw an error
# agar hum yaha per all work numeric data ke sath karenge tabhi yah work karega. otherwise yah hame error dega.
```

```python
var1.interpolate(method="linear",limit=1)
```

```python
# hum yaha per limit direction ka bhi use kar sakte hai.
print(var1)
var1.interpolate(limit_direction="forward",limit=1)            #limit_direction=forward,backward,both
```

```python
# limit area
# Iske throw hum bata sakte hai ki hame interpolation ko use karna hai ya nahi karna hai.
var1.interpolate(limit_area="inside")
```

```python
var1.interpolate(limit_area="outside")
#Agar hum orinal data ko change jarna chahte hai ,To hum yaha per inplace paramter ka bhi usse kar sakte hai.
```
